# Central Limit Theorem. 

A very much important theorem of infrential stastical probability, helps us to make predictions about the "Population data", based on the "Sample data". 

Random sampling of the fairly large population data in same size of n.

The mean of the sample data tends to teh mean of the population data. as n( size of the sample) increses. 

# Sample distribution 
random sampling of population data in same size, evey sample acts as a "Random_Varible". i.e. X1, X2, X3, X4,.. Xn.

#Expexted value of all the sample means. 
the avarage of all the sample means tends to #"mean value of the population data".

# Variance of all of the sampled means tends to variance of population over the size of smapled data. 
Irrespictive of the distribution of the population data, the disribution of the any sampled data approches to the "Normal_Distribution", and incresing n (i.e. the size of the population) the distribution curv(Bell shape) of samples becomes narrower, i.e. the standerd_deviation tends to low.  


# Simualtion of Central Limit Theorem. 
#using USA house prices data. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

df = pd.read_csv("C:\\Users\\HP\\Downloads\\USA_house_dataset.csv")
#print(df.describe)
#print("isnull:",len(df.isnull()))
#print(df.isnull())
#print(df.info())

sns.kdeplot(df["price"]) 
plt.title("Population distribution")
plt.show()  #skewed distributed population_data. (non-normal)

#taking sample of 110 house prices
samp = df["price"].sample(200) 
print("sample size:",len(samp))

sns.kdeplot(samp)  
plt.title("Sample_distribution")
plt.show
```

# Population distribution
here the data is left_skewed (i.e. non-normal).  

# Sample distribution of 200 house prices.
most likely to be a "Noraml distribution". 

```python
#maen and variences of population and sample data. 
popul = df["price"]

print("P_size",len(popul))

pmean = np.mean(popul)
print("population_mean",pmean)
print()

smean = np.mean(samp)
print("sample_mean",smean)
print()

if smean == pmean:
    print("S_mean and P_mean are approx same")
else:
    print("S_mean and P_mean are very much close.")
```

The spread of the sample data is tending to the variance of the population over size of the sample data

```python
pvar = np.var(popul)
print("poulation_spread:",pvar) 
print()
svar = np.var(samp)
print("sample_sperad:",svar)
print()
print("Sample_spread:",pvar/200) 
```

```python
#sample size of 200 -> 100 samples 
samples = [ ]

for i in range(100):
    samples.append(df["price"].sample(200)) 
       
print(f"sample size of each sample:",len(samples)) 
print()
samparray = np.array(samples) #Creating a araay of all 100 sapmles of size 200 
 
samp_dist = np.mean(samples,axis = 1)  #sample dsitribution 

print(f"Sample_distribution:",samp_dist)  #All 100 samples have the same mean like the mean of population data. 
print()
print("Every sample mean of the sample distribution is close to the mean of population data")

print()
Expval = np.mean(samp_dist)  #Mean of all the sampled means. 
print("Avarage of all sampled means:",Expval)

print()

if pmean == Expval:
    print("Expected val of sampled means house prices is same as the mean of population house prices")
else:
    print("Expected val of sampled means of house prices is almost same as the mean of population house prices")
```

```python
means = []

pmean = np.mean(popul)  #Population mean
#print("population_mean",pmean) 
means.append(pmean)

smean = np.mean(samp)   #Sample mean
#print("sample_mean",smean)
means.append(smean)

Expval = np.mean(samp_dist)  #Avg of all the sample means
#print("Avarage of all sampled means:",Expval)
means.append(Expval)

print(means)
```

# Confidenace interval 
#Emperical rule: 95% pobability that the mean of population data will lie between 2 standerd deviations of the sample 
distribution. 

```python
SD = np.std(samp_dist)/np.sqrt(200)  #Standerd Deviation

print("Standerd_diviation of Samp_distribution:",SD) 
print()

lower_limit = np.mean(samp_dist) - 2*SD 
upper_limit = np.mean(samp_dist) + 2*SD

print("Lower_lmit",lower_limit)
print("Upper_limit",upper_limit)   
print("Population_mean:",pmean)

print()
if lower_limit < pmean < upper_limit: 
    print("Population mean lies between 2 Standerd_Daviations")
```
